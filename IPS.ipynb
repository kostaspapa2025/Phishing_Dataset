```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Conv1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from sklearn.utils import class_weight


# Κατέβασμα του dataset από το GitHub
url = 'https://raw.githubusercontent.com/kdemertzis/EKPA/main/Data/pcap_data.csv'
data = pd.read_csv(url)

# Εκτύπωση των ονομάτων των στηλών για να βρεις το σωστό όνομα της στήλης των ετικετών
print(data.columns)

# Check if 'Class' column exists, if not, find the correct column name
if 'Class' not in data.columns:
    # Assuming the class column is the last one, adjust if necessary
    class_column_name = data.columns[-1]
    print(f"Using '{class_column_name}' as the class column name.")
else:
    class_column_name = 'Class'

# Προετοιμασία δεδομένων
data = data.dropna()  # Αφαίρεση κενών τιμών
X = data.drop(class_column_name, axis=1).values  # Επιλογή χαρακτηριστικών
y = data[class_column_name].values  # Ετικέτες

# Κωδικοποίηση ετικετών
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = tf.keras.utils.to_categorical(y)  # One-hot encoding

# Κανονικοποίηση των χαρακτηριστικών
X = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

# Διαχωρισμός δεδομένων σε train, validation, test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, stratify=y_train, random_state=42)

# Υπολογισμός class weights για την καταπολέμηση του class imbalance
unique_classes = np.unique(data[class_column_name])
num_classes = len(unique_classes)
# Convert unique_classes to numerical labels if they are not already
if isinstance(unique_classes[0], str):  # Check if labels are strings
    unique_classes_num = label_encoder.transform(unique_classes)
else:
    unique_classes_num = unique_classes
class_weights = class_weight.compute_class_weight(class_weight='balanced', classes=unique_classes_num, y=np.argmax(y_train, axis=1))
class_weights = {i: class_weights[i] for i in range(len(class_weights))}

# Κατασκευή μοντέλου Conv1D
model = Sequential([
    Conv1D(64, 3, activation='relu', input_shape=(X_train.shape[1], 1)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(y_train.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks
earlyStopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='min')
modelCheckPoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy', mode='max')

# Εκπαίδευση του μοντέλου
history = model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val),
                    class_weight=class_weights, callbacks=[earlyStopping, modelCheckPoint])

# Αξιολόγηση του μοντέλου
model.load_weights('best_model.h5') # Changed filename from 'best_model.hdf5' to 'best_model.h5'
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Accuracy on Test Data: {accuracy*100:.2f}%')

# Plotting Training and Validation Loss/Accuracy
plt.figure()
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.legend()

plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.legend()
plt.show()

# Αξιολόγηση του μοντέλου
model.load_weights('best_model.h5')
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Accuracy on Test Data: {accuracy*100:.2f}%')

# Plotting Training and Validation Loss/Accuracy
plt.figure()
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.legend()

plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.legend()
plt.show()


# Υποθέτουμε ότι η κλάση "0" = Normal, "1" = Attack
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# Κατώφλι εμπιστοσύνης για ενεργοποίηση δράσης
threshold = 0.85
attack_class_index = 1

print("IPS Simulation: Εντοπισμός και Αντίδραση σε επιθέσεις")

for i, prob in enumerate(y_pred_probs):
    predicted_class = np.argmax(prob)
    confidence = np.max(prob)

    if predicted_class == attack_class_index and confidence > threshold:
        print("Εντοπίστηκε πιθανή επίθεση (instance {i})")
        print("   ➤ Εμπιστοσύνη: {confidence*100:.2f}%")
        print("   ➤ Ενέργεια: Απόρριψη πακέτου / Ειδοποίηση διαχειριστή")
```
